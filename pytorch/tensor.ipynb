```python
import torch

print(torch.__version__)
```

```python
tensor1 = torch.tensor(10)

print(tensor1)
print(tensor1.size()) # torch.Size([])
print(tensor1.dtype)

tensor2 = torch.tensor([1, 2, 3, 4, 5])

print(tensor2)
print(tensor2.size()) # torch.Size([5])
print(tensor2.dtype) # torch.int64

tensor3 = torch.tensor([[1, 2, 3], [4, 5, 6]])

print(tensor3)
print(tensor3.size()) # torch.Size([2, 3])
print(tensor3.dtype) # torch.int64

tensor4 = torch.tensor([1.0, 2.0, 3.0])

print(tensor4)
print(tensor4.size())
print(tensor4.dtype) # torch.float32


```

```python
import numpy as np

tensor5 = torch.tensor(np.array([1, 2, 3, 4, 5]))

print(tensor5)
print(tensor5.size())
print(tensor5.dtype) # torch.int64


tensor6 = torch.tensor(np.array([1.0, 2.0, 3.0]))

print(tensor6)
print(tensor6.size())
print(tensor6.dtype) # torch.float64

tensor7 = torch.tensor(np.array([[1.0, 2, 3], [4, 5, 6]]))

print(tensor7)
print(tensor7.size())
print(tensor7.dtype) # torch.float64
```
